## Resolução das Tarefas Propostas

### Tarefa 1: Número de livros publicados após 1º de janeiro de 2000

```python
query = '''
SELECT COUNT(*) AS total_books
FROM books
WHERE publication_date > '2000-01-01'
'''

result = pd.read_sql(query, con=engine)
result
```

### Conclusão

A consulta identificou 819 livros publicados após 1º de janeiro de 2000.

Considerando que a base de dados contém 1.000 livros no total, observa-se que aproximadamente 81,9% do catálogo é composto por obras publicadas a partir dos anos 2000. Em contrapartida, apenas 181 livros foram publicados antes desse período.

Esses resultados indicam que a plataforma possui uma forte concentração de títulos contemporâneos, sugerindo um catálogo voltado principalmente para publicações mais recentes e alinhadas às tendências atuais do mercado editorial.

### Tarefa 2: Número de avaliações e classificação média para cada livro

O objetivo desta análise é identificar o nível de engajamento dos usuários com cada livro por meio da quantidade de avaliações recebidas, além de calcular a classificação média atribuída pelos leitores.

```python
query = '''
SELECT
    b.book_id,
    b.title,
    COUNT(DISTINCT rv.review_id) AS total_reviews,
    ROUND(AVG(rt.rating), 2) AS avg_rating
FROM books AS b
LEFT JOIN reviews AS rv
    ON b.book_id = rv.book_id
LEFT JOIN ratings AS rt
    ON b.book_id = rt.book_id
GROUP BY
    b.book_id,
    b.title
ORDER BY total_reviews DESC
'''

result = pd.read_sql(query, con=engine)
result.head(10)
```

### Observação dos Resultados

Entre os livros analisados, o maior número de resenhas identificado foi de 7, correspondente ao livro *Twilight (Twilight #1)*. Os demais títulos mais comentados apresentaram 6 resenhas cada, indicando uma diferença relativamente pequena entre os livros com maior engajamento.

Em relação às classificações médias, o livro *Harry Potter and the Prisoner of Azkaban* apresentou a maior avaliação entre os dez primeiros resultados (4,41), seguido por *The Book Thief* (4,26) e *The Glass Castle* (4,21).

### Conclusão

Os resultados sugerem que o engajamento dos leitores está distribuído de forma relativamente equilibrada entre os livros mais populares, sem a existência de um único título concentrando grande parte das resenhas.

Além disso, observa-se que os livros mais comentados apresentam avaliações médias elevadas, geralmente superiores a 4 pontos. Isso pode indicar uma relação positiva entre satisfação dos leitores e participação por meio de avaliações e resenhas, informação relevante para a identificação de obras com maior potencial de destaque dentro da plataforma.

### Tarefa 3: Identificar a editora que publicou o maior número de livros com mais de 50 páginas

Nesta etapa, busca-se identificar a editora responsável pela publicação do maior número de livros com mais de 50 páginas. O critério de páginas foi utilizado para excluir brochuras e publicações muito curtas, concentrando a análise em livros completos.

```python
query = '''
SELECT
    p.publisher,
    COUNT(b.book_id) AS total_books
FROM books AS b
INNER JOIN publishers AS p
    ON b.publisher_id = p.publisher_id
WHERE b.num_pages > 50
GROUP BY p.publisher
ORDER BY total_books DESC
LIMIT 5
'''

pd.read_sql(query, con=engine)
```

### Observação dos Resultados

Entre as editoras analisadas, a Penguin Books apresentou o maior número de livros com mais de 50 páginas, totalizando 42 publicações. A segunda colocada foi a Vintage, com 31 livros, resultando em uma diferença de 11 títulos entre as duas editoras.

Esse resultado indica uma liderança relativamente clara da Penguin Books dentro do catálogo analisado.

### Conclusão

A Penguin Books foi identificada como a editora com maior presença entre os livros completos da base de dados. A vantagem de 11 títulos em relação à segunda colocada sugere uma participação relevante da editora no catálogo disponível, contribuindo significativamente para a oferta de livros na plataforma.

### Tarefa 4: Identificar o autor com a maior classificação média

Nesta etapa, busca-se identificar o autor com a maior classificação média de seus livros.

Para garantir maior confiabilidade na análise, serão considerados apenas livros que receberam pelo menos 50 classificações dos usuários. Esse critério foi adotado para reduzir o impacto de livros com poucas avaliações, que poderiam distorcer os resultados e gerar conclusões menos representativas.

```python
query = '''
WITH books_50_ratings AS (
    SELECT
        book_id
    FROM ratings
    GROUP BY book_id
    HAVING COUNT(rating_id) >= 50
)

SELECT
    a.author,
    ROUND(AVG(r.rating), 2) AS avg_rating
FROM books b
INNER JOIN authors a
    ON b.author_id = a.author_id
INNER JOIN ratings r
    ON b.book_id = r.book_id
WHERE b.book_id IN (
    SELECT book_id
    FROM books_50_ratings
)
GROUP BY a.author
ORDER BY avg_rating DESC
LIMIT 10
'''

result = pd.read_sql(query, con=engine)
result
```

### Observação dos Resultados

Considerando apenas livros com pelo menos 50 classificações, o autor com a maior média de avaliação foi **J.K. Rowling/Mary GrandPré**, com nota média de **4,29**.

As posições seguintes foram ocupadas por **Markus Zusak/Cao Xuân Việt Khương** (4,26) e **J.R.R. Tolkien** (4,25). Observa-se que a diferença entre os três primeiros colocados é inferior a 0,05 pontos, indicando um nível de aprovação muito semelhante entre esses autores.

Além disso, todos os autores presentes no Top 10 apresentaram médias superiores a **3,75 pontos**, enquanto os cinco primeiros colocados superaram a marca de **4,0 pontos**. Esses resultados sugerem que os livros com maior volume de avaliações tendem a receber classificações positivas dos leitores.


### Conclusão

A análise identificou J.K. Rowling/Mary GrandPré como os autores com a maior classificação média entre os livros que receberam pelo menos 50 avaliações.

A adoção de um número mínimo de classificações aumentou a confiabilidade da análise, reduzindo possíveis distorções causadas por livros com poucas avaliações. Os resultados também sugerem que autores amplamente reconhecidos pelo público tendem a manter elevados níveis de satisfação entre os leitores, o que pode representar uma oportunidade para recomendações e estratégias de destaque dentro da plataforma.

### Tarefa 5: Número médio de avaliações entre usuários que classificaram mais de 50 livros

Nesta etapa, busca-se analisar o comportamento dos usuários mais ativos da plataforma. Para isso, serão considerados apenas os usuários que classificaram mais de 50 livros.

O objetivo é calcular o número médio de avaliações textuais (reviews) produzidas por esse grupo de usuários, permitindo compreender melhor seu nível de engajamento além das classificações numéricas.

```python
query = '''
WITH active_users AS (
    SELECT
        username
    FROM ratings
    GROUP BY username
    HAVING COUNT(book_id) > 50
)

SELECT
    ROUND(AVG(review_count), 2) AS avg_reviews
FROM (
    SELECT
        username,
        COUNT(review_id) AS review_count
    FROM reviews
    WHERE username IN (
        SELECT username
        FROM active_users
    )
    GROUP BY username
) t
'''

result = pd.read_sql(query, con=engine)
result
```

### Observação dos Resultados

A análise mostrou que os usuários que classificaram mais de 50 livros escreveram, em média, **24,33 avaliações textuais**.

Considerando que todos os usuários analisados atribuíram classificações a mais de 50 livros, observa-se que o número médio de resenhas é significativamente menor que o número de classificações realizadas. Isso indica que os usuários tendem a avaliar os livros com notas com maior frequência do que escrevem comentários detalhados sobre suas experiências de leitura.


### Conclusão

Os resultados sugerem que a participação dos usuários mais ativos ocorre principalmente por meio de classificações numéricas, enquanto a produção de avaliações textuais exige um nível maior de envolvimento.

Para a plataforma, esse comportamento indica uma oportunidade de desenvolver mecanismos que incentivem a criação de resenhas, como sistemas de recompensas, recomendações personalizadas ou recursos de interação social. O aumento do número de avaliações textuais pode enriquecer a experiência dos leitores e fornecer informações mais detalhadas sobre os livros disponíveis no catálogo.
